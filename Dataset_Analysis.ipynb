```python
# Dataset path list.
data = ["./DATA/s1/train_5core.tsv","./DATA/s2/train_5core.tsv","./DATA/s3/train_5core.tsv","./DATA/t1/train_5core.tsv","./DATA/t2/train_5core.tsv"]

for i in range(0,5):

    # Loading the first dataset then checking some imperfections, then cleaning it.
    df = pd.read_table(data[i])
    print("\nDATASET: {}".format(data[i][7:9]))
    print(df.isna().sum())
    print(df.duplicated().sum())
    df = df.drop_duplicates()

    # Total row.
    print("\nTotal row:")
    print(df.shape[0])

    # Item average per user.
    print("\nItem Average per User:")
    print(df.shape[0] / df["userId"].nunique())

    print("\nUnique User Id - Prdoduct Id Count:")
    print(str(df["userId"].nunique()) + "-" + str(df["itemId"].nunique()))

    # Matching datasets and checking if they have any user or product intersection.
    j = i + 1
    while j < 5:
        df2 = pd.read_table(data[j])
        print("\n{}-{} Number of userId's that intersect: {}".format(data[i][7:9], data[j][7:9], len(list(set(df["userId"].unique()) & set(df2["userId"].unique())))))
        print("{}-{} Number of itemId's that intersect: {}".format(data[i][7:9], data[j][7:9], len(list(set(df["itemId"].unique()) & set(df2["itemId"].unique())))))
        j += 1
```
